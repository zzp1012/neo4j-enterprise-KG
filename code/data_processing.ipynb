# Data Processing

## Dependency

````python
import math
import pandas as pd 
import numpy as np 
````

## Import Data

````python
company_information = pd.read_csv("./cleaned/company_infomation.csv")
company_information
````

````python
company_to_company = pd.read_csv("./cleaned/company_to_company.csv")
company_to_company
````

````python
person_to_company = pd.read_csv("./cleaned/person_to_company.csv")
person_to_company
````

````python
company_names = set(company_information["corporate_name"]) | set(company_to_company["company_name"]) | set(company_to_company["invested_company_name"]) | set(person_to_company["company_name"])
````

````python
person_to_company_split = person_to_company[person_to_company["investor"].isin(company_names)]
person_to_company_split
````

## NodeID

````python
companies = set(company_information["eeid"]).union(set(company_to_company["eeid"])).union(set(company_to_company["invested_company_eeid"])).union(set(person_to_company["eeid"]))
````

````python
company_part = pd.DataFrame(companies, columns = ["NodeID:ID(NodeID)"])
legal_person_info = company_information.dropna(subset = ["name_of_legal_person"])
legal_person_part = pd.DataFrame(np.arange(start= 0, stop= legal_person_info.shape[0],step=1), columns = ["NodeID:ID(NodeID)"])
investor_part = pd.DataFrame(np.arange(start= legal_person_info.shape[0], stop= legal_person_info.shape[0] + person_to_company.shape[0],step=1), columns = ["NodeID:ID(NodeID)"])
nodes = pd.concat([company_part, legal_person_part, investor_part], ignore_index=True)
nodes
````

````python
nodes.to_csv("./import_data/nodes.csv", index=False)
````

## Name Nodes

````python
names = set(company_information["corporate_name"]).union(set(company_to_company["company_name"])).union(set(company_to_company["invested_company_name"])).union(set(person_to_company["company_name"])).union(set(company_information["name_of_legal_person"])).union(set(person_to_company["investor"]))
````

````python
names.remove(np.nan)
````

````python
name_nodes = pd.DataFrame(names, columns = ["Name:ID(Name)"])
name_nodes
````

````python
name_nodes.to_csv("./import_data/name_nodes.csv", index=False)
````

## Type Nodes

````python
type_nodes = pd.DataFrame(set(person_to_company["stock_type"]), columns = ["Type:ID(Type)"])
type_nodes = type_nodes.dropna(subset=["Type:ID(Type)"])
type_nodes
````

````python
type_nodes.to_csv("./import_data/type_nodes.csv", index=False)
````

## City Nodes

````python
city_nodes = pd.DataFrame(set(company_information["city"]), columns = ["City:ID(City)"])
city_nodes = city_nodes.dropna(subset = ["City:ID(City)"])
city_nodes
````

````python
city_nodes.to_csv("./import_data/city_nodes.csv", index=False)
````

## Names Relationships

### Company Names

````python
company_names_part_1 = company_information[["eeid", "corporate_name"]]
company_names_part_1 = company_names_part_1.dropna(subset = ["corporate_name"])
company_names_part_1 = company_names_part_1.rename(columns = {"eeid":":END_ID(NodeID)", "corporate_name":":START_ID(Name)"})
company_names_part_1 = company_names_part_1[[":START_ID(Name)", ":END_ID(NodeID)"]]
company_names_part_1
````

````python
company_names_part_2 = company_to_company[["eeid", "company_name"]]
company_names_part_2 = company_names_part_2.dropna(subset = ["company_name"])
company_names_part_2 = company_names_part_2.rename(columns = {"eeid":":END_ID(NodeID)", "company_name":":START_ID(Name)"})
company_names_part_2 = company_names_part_2[[":START_ID(Name)", ":END_ID(NodeID)"]]
company_names_part_2
````

````python
company_names_part_3 = company_to_company[["invested_company_eeid", "invested_company_name"]]
company_names_part_3 = company_names_part_3.dropna(subset = ["invested_company_name"])
company_names_part_3 = company_names_part_3.rename(columns = {"invested_company_eeid":":END_ID(NodeID)", "invested_company_name":":START_ID(Name)"})
company_names_part_3 = company_names_part_3[[":START_ID(Name)", ":END_ID(NodeID)"]]
company_names_part_3
````

````python
company_names_part_4 = person_to_company[["eeid", "company_name"]]
company_names_part_4 = company_names_part_4.dropna(subset = ["company_name"])
company_names_part_4 = company_names_part_4.rename(columns = {"eeid":":END_ID(NodeID)", "company_name":":START_ID(Name)"})
company_names_part_4 = company_names_part_4[[":START_ID(Name)", ":END_ID(NodeID)"]]
company_names_part_4
````

````python
company_names = pd.concat([company_names_part_1, company_names_part_2, company_names_part_3, company_names_part_4], ignore_index=True)
company_names = company_names.drop_duplicates(keep='first', inplace=False)
company_names
````

````python
company_names.to_csv("./import_data/company_names.csv", index=False)
````

### Legal Person Names

````python
legal_person_names = pd.concat([legal_person_info.reset_index()["name_of_legal_person"], legal_person_part], axis = 1)
legal_person_names = legal_person_names.rename(columns = {"name_of_legal_person":":START_ID(Name)", "NodeID:ID(NodeID)":":END_ID(NodeID)"})
legal_person_names
````

````python
legal_person_names.to_csv("./import_data/legal_person_names.csv", index = False)
````

### Investor Names

````python
investor_names = pd.concat([investor_part,person_to_company["investor"].reset_index()["investor"]], axis = 1)
investor_names = investor_names.rename(columns = {"investor":":START_ID(Name)", "NodeID:ID(NodeID)":":END_ID(NodeID)"})
investor_names
````

````python
investor_names.to_csv("./import_data/investor_names.csv", index = False)
````

## Types Relationships

### Company Types

````python
company_types = pd.concat([pd.DataFrame(pd.Series('企业').repeat(company_part.shape[0])).reset_index()[0], company_part], axis = 1)
company_types = company_types.rename(columns = {0:":START_ID(Type)", "NodeID:ID(NodeID)":":END_ID(NodeID)"})
company_types
````

````python
company_types.to_csv("./import_data/company_types.csv", index = False)
````

### Legal Person Types

````python
legal_person_types = pd.concat([pd.DataFrame(pd.Series('自然人股东').repeat(legal_person_part.shape[0])).reset_index()[0], legal_person_part], axis = 1)
legal_person_types = legal_person_types.rename(columns = {0:":START_ID(Type)", "NodeID:ID(NodeID)":":END_ID(NodeID)"})
legal_person_types
````

````python
legal_person_types.to_csv("./import_data/legal_person_types.csv", index = False)
````

### Investor Types

````python
investor_types = pd.concat([person_to_company["stock_type"].reset_index()["stock_type"], investor_part], axis = 1)
investor_types = investor_types.rename(columns = {"stock_type":":START_ID(Type)", "NodeID:ID(NodeID)":":END_ID(NodeID)"})
investor_types = investor_types.dropna(subset = [":START_ID(Type)"])
investor_types
````

````python
investor_types.to_csv("./import_data/investor_types.csv", index = False)
````

## City Relationships

````python
cities = company_information[["city", "eeid"]]
cities = cities.rename(columns = {"city":":START_ID(City)", "eeid":":END_ID(NodeID)"})
cities = cities.dropna(subset = [":START_ID(City)"])
cities
````

````python
cities.to_csv("./import_data/cities.csv", index = False)
````

## Person to Company

````python
ptc_csv = person_to_company.reset_index()[["controlling_proportion", "subscription_amount", "eeid"]]
ptc_csv = pd.concat([investor_part, ptc_csv], axis = 1)
ptc_csv = ptc_csv.rename(columns = {"eeid":":END_ID(NodeID)", "subscription_amount":"value", "controlling_proportion":"ratio", "NodeID:ID(NodeID)":":START_ID(NodeID)"})
ptc_csv
````

````python
ptc_csv.to_csv("./import_data/person_to_company.csv", index=False)
````

## Company to Company

````python
ctc_csv = company_to_company[["eeid", "stock_percentage", "should_capi", "invested_company_eeid"]]
ctc_csv = ctc_csv.rename(columns = {"eeid":":START_ID(NodeID)", "invested_company_eeid":":END_ID(NodeID)", "stock_percentage":"ratio", "should_capi":"value"})
ctc_csv
````

````python
duplicated_names = set(person_to_company_split["investor"])
table_1 = company_information[company_information["corporate_name"].isin(duplicated_names)][["eeid", "corporate_name"]]
table_1 = table_1.rename(columns = {"corporate_name":"name"})
table_2 = company_to_company[company_to_company["company_name"].isin(duplicated_names)][["eeid", "company_name"]]
table_2 = table_2.rename(columns = {"company_name" : "name"})
table_3 = company_to_company[company_to_company["invested_company_name"].isin(duplicated_names)][["invested_company_eeid", "invested_company_name"]]
table_3 = table_3.rename(columns = {"invested_company_name" : "name", "invested_company_eeid": "eeid"})
table_4 = person_to_company[person_to_company["company_name"].isin(duplicated_names)][["eeid", "company_name"]]
table_4 = table_4.rename(columns = {"company_name" : "name"})
table = pd.concat([table_1, table_2, table_3, table_4], ignore_index=True)
table.drop_duplicates(subset = ["name"],keep='first', inplace = True)
table
````

````python
ctc_csv.to_csv('./import_data/company_to_company.csv', index=False) 
````

## Legal Person Relationships

````python
legal_persons = pd.concat([legal_person_part, legal_person_info.reset_index()["eeid"]], axis = 1)
legal_persons = legal_persons.rename(columns = {"NodeID:ID(NodeID)":":START_ID(NodeID)", "eeid":":END_ID(NodeID)"})
legal_persons
````

````python
legal_persons.to_csv("./import_data/legal_persons.csv", index = False)
````

```bash
cd ~
rm -rf neo4j-community-4.1.0
tar -xf neo4j-community-4.1.0-unix.tar.gz
cp neo4j.conf neo4j-community-4.1.0/conf/
cp data_upload/import_data/*.csv .
./neo4j-community-4.1.0/bin/neo4j-admin import --id-type=STRING --multiline-fields=true --nodes=NodeID=nodes.csv --nodes=Name=name_nodes.csv --nodes=Type=type_nodes.csv --nodes=City=city_nodes.csv --relationships=Invest=company_to_company.csv --relationships=Invest=person_to_company.csv --relationships=Is_name_of=company_names.csv --relationships=Is_name_of=investor_names.csv --relationships=Is_name_of=legal_person_names.csv --relationships=Is_type_of=company_types.csv --relationships=Is_type_of=investor_types.csv --relationships=Is_type_of=legal_person_types.csv --relationships=Is_city_of=cities.csv --relationships=Is_legal_person_of=legal_persons.csv
rm *.csv
rm import.report
```
